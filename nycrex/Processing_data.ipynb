```python
data['ppsf'] = data['price']/data['sqft']*12
data['ppsf'][np.isinf(data['ppsf'])] = 0
```

```python
data['new_bed'] = np.floor(data['bed'])
data['new_bed'] = [max(0,min(x,5)) for x in data['new_bed']]
data.groupby(['boro','new_bed'])[['price','daysMkt']].apply(lambda x: np.round(np.mean(x)))
```

```python
# price per zip:
data.groupby(['zip'])[['price','daysMkt']].apply(lambda x: np.round(np.mean(x)))
```

```python
#grouping amenities in categories:
data['lam_outdoor']     = data['lAm_balcony'].astype(int)     +  data['lAm_courtyard'].astype(int) + \
                          data['lAm_garden'].astype(int)      +  data['lAm_patio'].astype(int)     + \
                          data['lAm_roofDeck'].astype(int)    +  data['lAm_terrace'].astype(int)

data['lam_luxury']      = data['lAm_centAC'].astype(int)      +  data['lAm_concier'].astype(int)   + \
                          data['lAm_liveInSuper'].astype(int) +  data['lAm_doorman'].astype(int)   + \
                          data['lAm_golf'].astype(int)        +  data['lAm_fireplace'].astype(int) + \
                          data['lAm_tennis'].astype(int)      +  data['lAm_pool'].astype(int)      + \
                          data['lAm_loft'].astype(int)

data['lam_mnties']      = data['lAm_gym'].astype(int)         +  data['lAm_laundry'].astype(int)   + \
                          data['lAm_washerDryer'].astype(int) +  data['lAm_lounge'].astype(int)    + \
                          data['lAm_mediaRm'].astype(int)     +  data['lAm_fios'].astype(int)      + \
                          data['lAm_dishwasher'].astype(int)  +  data['lAm_recFac'].astype(int)

data['lam_convnce']     = data['lAm_bikeRm'].astype(int)      +  data['lAm_basement'].astype(int)  + \
                          data['lAm_gatedCom'].astype(int)    +  data['lAm_elevator'].astype(int)  + \
                          data['lAm_packageRm'].astype(int)   +  \
                          data['lAm_childPlay'].astype(int)   +  data['lAm_smokeFree'].astype(int) + \
                          data['lAm_storage'].astype(int)

data['lam_car']         = data['lAm_garage'].astype(int)      +  data['lAm_parking'].astype(int)

data['lam_guarantorOK'] = data['lAm_guarantorOK'].astype(int)

data['lam_furnished']   = data['lAm_furnished'].astype(int)

data['lam_pets']        = data['lAm_pets'].astype(int)

data['lam_sublet']      = data['lAm_sublet'].astype(int)
```

```python
data['new_bath'] = [max(0,min(x,5)) for x in data['bath']]
```

```python
#data.replace(u'\xb7', '').replace(u'\u2022', '').replace(u'\xb2', '')
data.to_csv('app/data/listings.csv', encoding='utf-8')
```
